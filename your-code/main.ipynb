## 🔹 Collinearity Analysis for Logistic Regression

Logistic regression assumes independence between predictors.  
When predictors are strongly correlated (collinear), the model struggles to estimate which one actually influences the target.  
This leads to:
- **Unstable coefficients**
- **Inflated standard errors**
- **Misleading interpretations**

### 1. Packet-related features
- `SOURCE_APP_PACKETS` ↔ `APP_PACKETS` → correlation ≈ **1.00**  
- `SOURCE_APP_PACKETS` ↔ `REMOTE_APP_PACKETS` → correlation ≈ **0.99**  
- `APP_PACKETS` ↔ `REMOTE_APP_PACKETS` → correlation ≈ **0.99**

👉 These three features carry almost identical information.  
**Action:** Keep **one** (e.g., `SOURCE_APP_PACKETS`) and drop the other two.

---

### 2. Byte-related features
- `SOURCE_APP_BYTES` ↔ `APP_BYTES` → correlation ≈ **0.87**  
- `REMOTE_APP_BYTES` ↔ `APP_BYTES` → correlation ≈ **0.78**  
- `SOURCE_APP_BYTES` ↔ `REMOTE_APP_BYTES` → correlation ≈ **0.88**

👉 These features are also highly correlated and largely redundant.  
**Action:** Keep **one** (e.g., `APP_BYTES`) and drop the others.

---

### 3. Other variables
- Features such as `URL_LENGTH`, `NUMBER_SPECIAL_CHARACTERS`, and `CONTENT_LENGTH` show low correlations (< 0.3).  
- These provide unique information and are safe to keep.

---

## ✅ Decision
To reduce multicollinearity:
- **Drop:** `APP_PACKETS`, `REMOTE_APP_PACKETS`, `SOURCE_APP_BYTES`, `REMOTE_APP_BYTES`  
- **Keep:** `SOURCE_APP_PACKETS` (packets representative), `APP_BYTES` (bytes representative), and all other low-correlation features  

---

### 📌 Final Note
> We observed very strong collinearity among packet-related features (`SOURCE_APP_PACKETS`, `REMOTE_APP_PACKETS`, `APP_PACKETS`) and byte-related features (`SOURCE_APP_BYTES`, `REMOTE_APP_BYTES`, `APP_BYTES`).  
> In logistic regression, this multicollinearity can cause unstable coefficients and poor interpretability.  
> To address this, we retained only one representative feature from each group (`SOURCE_APP_PACKETS` and `APP_BYTES`) and dropped the redundant ones.  
> This reduces redundancy while preserving predictive power.

Since we have fixed the country values, can we convert this column to ordinal now?

Not yet. If you reflect on the previous labs how we handle categorical columns, you probably remember we ended up dropping a lot of those columns because there are too many unique values. Too many unique values in a column is not desirable in machine learning because it makes prediction inaccurate. But there are workarounds under certain conditions. One of the fixable conditions is:

#### If a limited number of values account for the majority of data, we can retain these top values and re-label all other rare values.

The `WHOIS_COUNTRY` column happens to be this case. You can verify it by print a bar chart of the `value_counts` in the next cell to verify:

```python
def print_bar_plot(x,y):
    plt.bar(x, y)
    plt.figure(figsize=(15,15))
    return plt.show()
```

```python
print_bar_plot(websites.WHOIS_COUNTRY.unique(),websites.WHOIS_COUNTRY.value_counts());
```

#### After verifying, now let's keep the top 10 values of the column and re-label other columns with `OTHER`.

```python
# Your code here
# Get top 10 most frequent values
top_10 = websites["WHOIS_COUNTRY"].value_counts().nlargest(10).index

# Replace all other values with "OTHER"
websites["WHOIS_COUNTRY"] = websites["WHOIS_COUNTRY"].apply(lambda x: x if x in top_10 else "OTHER")

# Verify
print(websites["WHOIS_COUNTRY"].value_counts())
```

Now since `WHOIS_COUNTRY` has been re-labelled, we don't need `WHOIS_STATEPRO` any more because the values of the states or provinces may not be relevant any more. We'll drop this column.

Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
top_3 = ["Microsoft", "Apache", "nginx"]
#I created the function here in order to loop through the unique server values. Each value contains the names Microsoft, Apache, or nginx. The categorize_server function first loops through a pd to check for nan values or values that dont contain these 3 categories and returns Other. The for loop checks for the names of each variable and converts to lower for easy matching reference. It then returns the matching server name and this is applied to the websites dataframe
def categorize_server(x):
    if pd.isna(x):  
        return "Other"
    for server in top_3:
        if server.lower() in x.lower():  
            return server
    return "Other"

websites["SERVER"] = websites["SERVER"].apply(categorize_server)

# Check unique values
print(websites["SERVER"].unique())
```
